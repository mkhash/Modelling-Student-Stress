- extracting user id's

```python
path = './ema_data/Activity/Activity_u08.json'
path1 = './ema_data/Activity'

yars  = [pos_json for pos_json in os.listdir(path1) if pos_json.endswith('.json')]

uids = np.array([re.findall(r'\d{2}', i)[0] for i in yars])
uids
```

```python
yar = pd.read_json(path).head()
yar['uid'] = 1
yar.set_index('uid', inplace = True)
```

```python
def folder_to_df(name_of_folder):

    # this finds our json files
    path_to_json = './ema_data/' + name_of_folder
    json_files = [pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json')]
    
    
    #instantiate array for each student's responses
    responses = pd.DataFrame()
    
    # we need both the json and an index number so use enumerate()
    for index, file_name in enumerate(json_files):
        
        file_path = path_to_json + '/' + file_name
        user_df = pd.read_json(file_path)
        
        #set index to user's id and reset index
        user_df['uid'] = uids[index]
        user_df.set_index('uid', inplace = True)
        user_df.reset_index(inplace = True)
        #user_df.sort_values(['resp_time'], inplace = True)
        
        responses = responses.append(user_df, sort = False)
        
    # return array of responses
    return responses
```

create_key_table:
- input: name of key (string)
- output: data frame with column names as index and actual response as value

```python
def visualize_key(key):
    return pd.DataFrame.from_dict(key['questions']).iloc[:,[1,2,0]].set_index('question_id')
```

```python
def create_key_table(key_name):
    with open('ema_keys/' + key_name + '.txt') as json_file:
        data = json.load(json_file)
    return visualize_key(data)
```

The location variables are actually strings and not lists of [lat, long] representation so I'll create a function that converts them, and apply it to all of our datasets.

```python
str_resp.head()
```

```python
def convert_to_num(df):
    lat = []
    long = []
    
    for i in df['location'].astype('str'):
        #print(i)
        if i == 'Unknown' or len(i) < 3 or i == 'nan' or i == 'null':
            lat.append(float('nan'))
            long.append(float('nan'))
        else:
            first = d.Decimal(str.split(i, ',')[0])
            second = d.Decimal(str.split(i, ',')[1])
            lat.append(first)
            long.append(second)
    df['lat'] = lat
    df['long'] = long
```

```python
def generate_colors(n):
  rgb_values = []
  r = int(random.random() * 256)
  g = int(random.random() * 256)
  b = int(random.random() * 256)
  step = 256 / n
  for _ in range(n):
    r += step
    g += step
    b += step
    r = int(r) % 256
    g = int(g) % 256
    b = int(b) % 256
    rgb_values.append((r,g,b))
  return rgb_values
```
